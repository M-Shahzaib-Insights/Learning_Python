# Adding more members to an existing list
- append(): appends/adds a member at very end/last
- insert(): add one value at any specified location.
- extend(): add one or list of values at the end of the list
- (+) : add the lists together

```python
#append(): appends/adds a member at very end/last (Only one value)
students.append("Rashid")
```

```python
students
```

```python
students.append("Rashid","Qasim")
```

```python
students.insert(3,"Sajid")
```

```python
students
```

```python
students.insert(100,"Mani")
students
```

```python
staff = ["Wahid", "Kamil", "Jami"]
```

```python
students.append(staff)
students
```

```python
students.extend(staff)
```

```python
students
```

```python
staff.extend("Hello")
staff
```

```python
students + staff
```

# Deleting a member from existing list
- del statement (Universal function for whole python): deletes permanently via index
- remove(): deletes parmanently via value
- pop(): deletes paramanently from last index
-        can delete a member from given index
-        it returns deleted value
- clear(): clears the complete list

```python
del students[4]
students
```

```python
del students[11]
students
```

```python
students.remove("Mani")
students
```

```python
print (students)
students.pop
students
```

```python
students.pop(5)
```

```python
del students[3:20]
students
```

```python
students.clear()
students
```

```python
students.extend('2ab5h')
students
```

```python
students.sort()
students
```
